```python
import numpy as np
import tensorflow as tf
```

```python
sess = tf.Session()
w1 = tf.Variable([0.3])
w2 = tf.Variable([-0.1])
tt = tf.constant(0.2)
ct = tf.constant(0.5)
x1 = tf.placeholder(tf.float32,shape=1,name='x2-input')
x2 = tf.placeholder(tf.float32,shape=1,name='x2-input')
yd = tf.placeholder(tf.float32,shape=1,name='y-real')

cal_y = tf.sign(w1*x1+w2*x2-tt)*ct+ct
delta = yd - cal_y

#train_step = tf.train.AdamOptimizer(0.1).minimize(delta)

alpha = tf.constant(0.1)
change_w1 = tf.assign(w1,w1+alpha*x1*delta)
change_w2 = tf.assign(w2,w2+alpha*x2*delta)
train_step = [change_w1,change_w2]

init_op = tf.global_variables_initializer()
```

```python
steps = 10
```

```python
sess.run(init_op)

d1 = {x1:[0.0],x2:[0.0],yd:[0.0]}
d2 = {x1:[0.0],x2:[1.0],yd:[0.0]}
d3 = {x1:[1.0],x2:[0.0],yd:[0.0]}
d4 = {x1:[1.0],x2:[1.0],yd:[1.0]}

count = 0
def loop(count):
    print("count, ", count)
    for d in [d1,d2,d3,d4]:
        print("w1: ", sess.run(w1,feed_dict=d))
        print("w2: ", sess.run(w2,feed_dict=d))
        print("cal_y: ", sess.run(cal_y,feed_dict=d))
        print("delta: ", sess.run(delta,feed_dict=d))
        sess.run(train_step,feed_dict=d)

        print("********")
        print("new_w1: ", sess.run(w1,feed_dict=d))
        print("new_w2: ", sess.run(w2,feed_dict=d))
        print("")
        print("-----------")
        print("")
```

```python
for i in range(5):
    loop(i)
```
